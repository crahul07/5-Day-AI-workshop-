```python
Climate_data = [
    {"City": "City A", "Temperature": 25, "Carbon_Footprint": 500},
    {"City": "City B", "Temperature": 35, "Carbon_Footprint": 450},
    {"City": "City C", "Temperature": 22, "Carbon_Footprint": 550},
    {"City": "City D", "Temperature": 20, "Carbon_Footprint": 480},

]

high_tem_threshold = 26

# Store full dictionary instead of just names
high_tem_cities = [city for city in Climate_data if city["Temperature"] > high_tem_threshold]

print("Cities with High Temperature (>26°C):")

for city in high_tem_cities:
    print(f"{city['City']} - ({city['Temperature']}°C)")
```

```python
import pandas as pd
import numpy as np

data = {
    "Energy Source": ["Solar", "Wind", "Hydro", "Nuclear", "Coal", "Biogas"],
    "Energy Consumption (MWh)": [1000, 800, np.nan, 1200, np.nan, 900],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}

engery_df = pd.DataFrame(data)

print("Original Data With Missing Values:")
print(engery_df)

print("\nHead Values:\n")
print(engery_df.head(3))

print("\nTail Values:\n")
print(engery_df.tail())

print("\nDescribe Values:\n")
print(engery_df.describe())
```
