```python
energy_values = [100, 200, 300, 400, 500]
carbon_emission = [10, 20, 30, 40, 50]

def_reg = pd.DataFrame({'Energy_Consumption': energy_values, 'Carbon Emission': carbon_emission})
sns.regplot(x='Energy_Consumption', y='Carbon Emission', data=def_reg)
plt.title('Energy Consumption vs. Carbon Emission')
plt.xlabel('Energy Consumption')
plt.ylabel('Carbon Emission')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/appliance_energy.csv')
print(df.head())
print(df.shape)
print(df.info())
print(df.describe)
```

```python
X=df[['Temperature (°C)']]
Y=df['Energy Consumption (kWh)']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=2,random_state=2)
model=LinearRegression()
model.fit(X_train,Y_train)
X_train.shape
X_test.shape
Y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

r2 = r2_score(Y_test, Y_pred)
print("R-squared:", r2)

```

```python
plt.scatter(X_test, Y_test, color='blue', label='Test Data')  # Actual data
plt.plot(X_test, Y_pred, color='red', label='Regression Line')  # Predicted data
plt.xlabel('Temperature (°C)')
plt.ylabel('Energy Consumption (kWh)')
plt.legend()
plt.title('Energy Consumption Prediction using Simple Linear Regression')
plt.show()
```
