```python
import numpy as pd
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

```

```python
X,y = make_moons(n_samples=300,noise=0.05,random_state=0)
X_scaled = StandardScaler().fit_transform(X)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],s=20)
plt.title("Origanal Data Distribution")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.grid(True)
#
```

```python
kmean = KMeans(n_clusters=2,random_state=0)
clusters = kmean.fit_predict(X_scaled)
```

```python
#Plot K-Means clustered data
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', s=30)
plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
db = DBSCAN(eps=0.3,min_samples=5)
clusters = db.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', s=30)
plt.title("DBSCAN Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
